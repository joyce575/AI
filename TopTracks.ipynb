# Question: How many pop songs were most popular in the US tracks in 2024 compared to the Global tracks? 
I will be conducting an observational study based on the whole population. The data has already been recorded, and as a researcher will only need to observe the data and also to answer my question. I will need to examine the entire population, not just a sample.
 Ms Spindt collected the data. She used AI to list out for her, and then she wrote the wiki genre by hand. Some biases that might occur in this data are the genres' popularity and how they might be overrepresented.

```python
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------------------------
# 1. Load both datasets
# -------------------------------------------------
usa = pd.read_csv("usa.csv")
global_df = pd.read_csv("global.csv")

# Add labels if not present
usa["Playlist"] = "USA"
global_df["Playlist"] = "Global"

# -------------------------------------------------
# 2. Combine the two datasets
# -------------------------------------------------
df = pd.concat([usa, global_df], ignore_index=True)

# Standardize the Playlist column (just in case)
df["Playlist"] = df["Playlist"].str.strip().str.title()

# -------------------------------------------------
# 3. Filter for POP songs (AI Genre contains "Pop")
# -------------------------------------------------
pop_df = df[df["AI Genre"].str.contains("Pop", case=False, na=False)]

# -------------------------------------------------
# 4. Create frequency table
# -------------------------------------------------
freq_table = pop_df["Playlist"].value_counts()
print("Frequency Table of Pop Songs:")
print(freq_table)

# -------------------------------------------------
# 5. Bar chart (2 bars)
# -------------------------------------------------
plt.figure(figsize=(7,5))
plt.bar(freq_table.index, freq_table.values)
plt.xlabel("Playlist")
plt.ylabel("Number of Pop Songs")
plt.title("Number of Pop Songs in USA vs Global Tracks (2024)")
plt.tight_layout()
plt.show()
```

# Conclusion: In 2024, the Global playlist had 21 pop songs, while the USA playlist had 18. This means pop songs were slightly more popular globally than in the U.S.
