```python
import gymnasium as gym

# Create Blackjack environment (text-based)
env = gym.make('Blackjack-v1', render_mode="human")

n_games = 100
hands_per_game = 10
games_with_6_wins = 0

for game_num in range(1, n_games + 1):
    wins = 0
    
    for hand_num in range(1, hands_per_game + 1):
        done = False
        observation, info = env.reset()
        
        while not done:
            player_sum, dealer_card, usable_ace = observation
            
            # Simple policy: hit if sum < 17, else stick
            if player_sum < 17:
                action = 1  # hit
            else:
                action = 0  # stick
            
            observation, reward, terminated, truncated, info = env.step(action)
            done = terminated or truncated
        
        if reward > 0:
            wins += 1
    
    if wins == 6:
        games_with_6_wins += 1
    
    print(f"Game {game_num}: {wins} wins")

env.close()

print(f"\nOut of {n_games} games, {games_with_6_wins} games had exactly 6 wins.")
print(f"Empirical probability of winning exactly 6 hands: {games_with_6_wins/n_games:.4f}")
```

# Conclusion: In 100 Blackjack games of 10 hands, the probability of winning exactly 6 hands was about 0.19, matching closely with the theoretical probability of 0.20, because the simulation uses enough games for the empirical results to approximate the expected binomial outcomes.
