```python
import pandas as pd 
import random
from collections import Counter
import matplotlib.pyplot as plt
```

```python
roll = random.randint(1, 6)
print(roll)   # gives a number between 1 and 6
```

```python
trials = 100
results = []

for _ in range(trials):
    roll = random.randint(1, 6)
    results.append(roll)

print(results[:100])  # shows 100 results
```

```python
counts = Counter(results)

# Step 3: prepare data for plotting
sides = list(range(1, 7))  # sides of the die (1–6)
frequencies = [counts.get(side, 0) for side in sides]

# Step 4: plot bar chart
plt.bar(sides, frequencies, color="skyblue", edgecolor="black")
plt.xlabel("Die Face")
plt.ylabel("Frequency")
plt.title("Die Roll Simulation (100 Trials)")
plt.xticks(sides)  # ensure all sides show up
plt.show()
```

```python
N = 9     # number of sides (change this to whatever you want)
trials = 100 # number of rolls

# Step 1: simulate die rolls
results = [random.randint(1, N) for _ in range(trials)]

# Step 2: count frequency of each outcome
counts = Counter(results)

# Step 3: prepare data for plotting
sides = list(range(1, N + 1))  # die faces from 1 to N
frequencies = [counts.get(side, 0) for side in sides]

# Step 4: generate a color for each bar
colors = plt.cm.tab20.colors  # built-in colormap (20 distinct colors)
bar_colors = [colors[i % len(colors)] for i in range(N)]

# Step 5: plot bar chart
plt.bar(sides, frequencies, color=bar_colors, edgecolor="red")
plt.xlabel("Die Face")
plt.ylabel("Frequency")
plt.title(f"{N}-Sided Die Simulation ({trials} Trials)")
plt.xticks(sides)
bars = plt.bar(sides, frequencies, color=bar_colors, edgecolor="black")


# Step 6: add frequency labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,  # x position (center of bar)
        height + 0.3,                     # y position (a little above bar)
        str(height),                      # label = frequency value
        ha='center', va='bottom'
    )

plt.show()
```
